### 08. Bar-Graph

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# 한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 윈도우 전용
matplotlib.rcParams['font.size'] = 15 #폰트 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['Kang', 'Seo', 'Jung'] # 이름
values = [190, 187, 184] # 키

plt.bar(labels, values)
```

```python
labels = ['Kang', 'Seo', 'Jung'] # 이름
values = [190, 187, 184] # 키

plt.bar(labels, values, color='r')
```

```python
labels = ['Kang', 'Seo', 'Jung'] # 이름
values = [190, 187, 184] # 키
colors=['r','g','b'] # plt를 선언할때 색을 지정 할 수 있지만, 이렇게 변수화도 가능하다. 


plt.bar(labels, values, color = colors, alpha=0.3)
```

```python
labels = ['Kang', 'Seo', 'Jung'] # 이름
values = [190, 187, 184] # 키

plt.bar(labels, values)
plt.ylim(175, 195) # ylim(시작값, 끝값) 으로 밸류값 조정 가능 
```

```python
plt.bar(labels, values, width=0.5) # width = 바 굵기 조절 가능 
```

```python
# Bar굵기 및 x축 텍스트 기울기
plt.bar(labels, values, width=0.5)
plt.xticks(rotation=45) #축의 텍스트를 기울 수 있음
plt.yticks(rotation=45);
```

```python
#x축에 들어갈 값 바꾸는 방법
ticks = ['No.1', 'No.2', 'No.3'] #번호
labels = ['Kang', 'Seo', 'Jung'] # 이름
values = [190, 187, 184] # 키

plt.bar(labels, values)
plt.xticks(labels,ticks);
```
