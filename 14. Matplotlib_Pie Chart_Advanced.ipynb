### 14. Pie Chart Advanced

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# 한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 윈도우 전용
matplotlib.rcParams['font.size'] = 15 #폰트 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30,25,20,13,10,2]
labels = ['Python','Java','JavaScript','C#','C','ETC']
colors = ['#ffadad','#ffd6a6','#fdffb6','#caffbf','#9bf6ff','#a0c4ff'] # 색상 변경
explode = [0.05] * 6

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90,
        counterclock=False, colors=colors, explode=explode);
```

```python
# wedgeprops을 써서 파이 차트를 꾸밀 수 있다
wedgeprops={'width':0.6, 'edgecolor':'w', 'linewidth':3}
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90,
        counterclock=False, colors=colors, wedgeprops=wedgeprops);
```

```python
# 텍스트 제어
# pctdistance='' 파이 차이의 텍스트 위치 조정

def custom_autopct(pct):
    return ('%.0f%%' % pct) if pct >= 10 else ''
    #return '{:.1f}%'.format(pct) if pct >= 10 else '' 와 동일
    # {:.1f%} -> .1f 소수점 한자리까지 보여주고 + %

plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90,
        counterclock=False, colors=colors, wedgeprops=wedgeprops,
       pctdistance=0.7);
```
