### 04. Style

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# 한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 윈도우 전용
matplotlib.rcParams['font.size'] = 15 #폰트 크기
matplotlib.rcParams['axes.unicode_minus'] = False

x = [1,2,3]
y = [2,4,8]
```

### Line and Marker

```python
plt.plot(x,y);
```

```python
# 선을 굵게
plt.plot(x,y, linewidth=3);
```

```python
# 동그라미로 시작, 꺾이는 부분, 끝 표시 가능
plt.plot(x,y, marker='x'); # o,x,v 사용 가능, 
```

```python
# 라인을 없애고 표시만 남기기
plt.plot(x,y,marker='v',linestyle='None', markersize=10); #'markersize로 크기 조절 가능'
```

```python
# 마커 색상 변경
plt.plot(x,y,marker='v',markersize=10, markeredgecolor='red'); #'markeredgecolor로 색상 변경
```

```python
# 마커 안의 색상 변경
plt.plot(x,y,marker='v',markersize=10, markeredgecolor='red', markerfacecolor='yellow'); #'markerfacecolor 색상 변경
```

```python
# 실선
plt.plot(x,y,linestyle=':');
```

```python
plt.plot(x,y, linestyle='--')
plt.plot(np.random.rand(10), color='gold')
plt.plot(np.random.rand(10), 'g:')
plt.plot(np.random.rand(10), color='#1234FF')
```

```python
# 스타일 포맷으로 넘기기
plt.plot(x,y, 'ro--'); # 'r(color)' / 'o(marker)' / '--(linestyle)'
```

```python
plt.plot(x,y, 'bv:');
```

```python
plt.plot(x,y, 'go');
```

```python
# 마커 축약어
plt.plot(x,y,marker='o', mfc='red', ms=10, mec='blue', ls=':');
# mfc = markerfacecolor
# ms = markersize
# mec = markeredgecolor
# ls = linestyle
```

```python
#투명도
plt.plot(x,y,marker='o', mfc='red', ms=10, mec='blue', ls=':', alpha=0.3); # alpha로 투명도 조절
```

```python
# 그래프 크기
plt.figure(figsize=(10, 5)) # 'figsize=(넓이, 높이)
plt.plot(x,y);
```

```python
plt.figure(figsize=(10, 5), dpi=200)
plt.plot(x,y);
```

```python
# 그래프 배경색
plt.figure(facecolor='yellow')
plt.plot(x,y)
```
