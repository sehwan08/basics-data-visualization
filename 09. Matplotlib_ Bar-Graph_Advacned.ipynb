### 09. Bar-Graph Advanced

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# 한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 윈도우 전용
matplotlib.rcParams['font.size'] = 15 #폰트 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['Kang', 'Seo', 'Jung'] # 이름
values = [190, 187, 184] # 키

# 옆으로 누은 막대그래프
plt.barh(labels, values)
plt.xlim(175, 195) #값 조절
```

```python
# 막대 꾸미기
bar = plt.bar(labels, values)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

```python
# 막대에 글자 넣기
bar = plt.bar(labels, values)
plt.ylim(175, 195)



for idx, rect in enumerate(bar):
    print(rect)
    print(rect.get_width())
    plt.text(idx, rect.get_height()+0.5, values[idx], ha='center', color='r')
```
